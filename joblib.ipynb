```python
import joblib
```

```python
from sklearn.datasets import make_classification
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X,y=make_classification(n_samples=100,n_features=10,n_informative=2)
```

```python
X
```

```python
y
```

```python
model=LogisticRegression()
```

```python
model.fit(X,y)
```

```python
model.get_params
```

```python
X_new,_=make_classification(n_samples=5,n_features=10,n_informative=2)
```

```python
X_new
```

```python
predictions=model.predict(X_new)
```

```python
joblib.dump(model,"LogisticReression.pkl")
```

```python
loaded_model=joblib.load("LogisticReression.pkl")
```

```python
new_data,_=make_classification(n_samples=2,n_features=10,n_informative=2)
```

```python
loaded_model.predict(new_data)
```
