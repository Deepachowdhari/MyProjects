```python
#resizing an image using pillow library

from PIL import Image
img=Image.open('/content/flower.jpg')
img_resized=img.resize((300,300))
```

```python
img_resized.save('flower_resized.jpg')
```

```python
img_resized=mpimg.imread("/content/flower_resized.jpg")
img_plot=plt.imshow(img_resized)
plt.show()
```

```python
#RGB to Greyscale conversion using OpenCV
import cv2
img=-cv2.imread('/content/flower.jpg')

```

```python
type(img)
```

```python
img.shape
```

```python
grey_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grey_img)
```
