#                              # Central Limit Theorem

#Central limit theorem states that if we take many samples from any 
population distribution, the distribution of sample means will appraoch a normal distribution.

#Central Limit Theorem
1.Start with a Population
    a.In this notebook will take a look on Right-skewe distribution,Binomial distribution and Poisson distribution
2.Take samples from the distribution
    a.ideally more than 25 numbers for a sample
    b.Needs to be atleast 30 samples
3.Find the mean of each sample
4.Plot the means on a Histogram
5.More samples=closer to a Normal Distribution

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(13)
```

```python
#Example 1: Skewed Distribution
```

```python
population_size=100000
```

```python
scale=2
```

```python
skewed=np.random.exponential(scale=scale,size=population_size)
```

```python
sns.histplot(skewed,kde=True)
plt.title("Right Skewed Data")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```

```python
skewed_means=[]
num_samples=1000
```

```python
sample_size_v1=30
```

```python
for i in range(num_samples):
    sample=np.random.choice(skewed,size=sample_size_v1,replace=False)
    sample_mean=sum(sample)/sample_size_v1
    skewed_means.append(sample_mean)
```

```python
plt.hist(skewed_means,bins=30,edgecolor='b',density=True)
plt.title("Right Skewed means")
plt.xlabel("sample means")
plt.ylabel("frequency")
plt.show()
```

```python
#Increasing the sample size
```

```python
sample_size_v2=100
```

```python
for i in range(num_samples):
    sample=np.random.choice(skewed,size=sample_size_v2,replace=False)
    sample_mean=sum(sample)/sample_size_v2
    skewed_means.append(sample_mean)
```

```python
plt.hist(skewed_means,bins=30,edgecolor='b',density=True)
plt.title("Right Skewed means")
plt.xlabel("sample means")
plt.ylabel("frequency")
plt.show()
```
